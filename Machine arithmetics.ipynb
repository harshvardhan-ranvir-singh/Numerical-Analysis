### A simple worked example

```python
# Here I'm requesting floating point real number
2.0 / 3
```

This is not exactly two third, this is clearly reasonably close to two third but that's not exactly two third.
Now, what is this number and how close is close? That is decided by the computer not us.

```python
# 1e20 is scientific notation for 10 to the power 20. Now we expect the calculation to be equal 1.
1e20 + 1 - 1e20
```

Huh! what's just happened. Is it wrong?

```python
from math import sqrt

def func(x):
    num = sqrt(1+x) + sqrt(x)
    den = sqrt(1+x) - sqrt(x)
    return 1./den - num

# we expect the result to be zero
```

```python
for x in [1e-20, 1, 1e3, 1e10, 1e20]:
    print(x, "  -->  ", func(x))
```
